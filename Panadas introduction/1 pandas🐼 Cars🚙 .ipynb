DATA STRUCTURES SUPPORTS IN PANDA

1.SERIES (only 1 column supports)

2.DATA FRAME (its look like table's supports number of columns and rows)

3.PANEL(its normally uses for server side)

```python
#series creation
myseries=pd.Series([1,"kannan","bpo",28,"Bangalore"]) #but can't give another list in same series only support 1 list
print(myseries)
```

```python
#call specific data with index
print(myseries[4])#to call bangalore 
```

```python
# how to give index specific index
myseries=pd.Series([1,"kannan","bpo",28,"Bangalore"],index=[1,2,3,4,5])
print(myseries)
print(myseries[3])
```

```python
#example 2
myseries=pd.Series([1,"kannan","bpo",28,"Bangalore"],index=["a","b","c","d","e"])
print(myseries)
print(myseries["c"])
```

```python
#for creating Data frame
mydict={'no':1,'name':"kannan",'age':28,'place':"kollam"}
df=pd.DataFrame(mydict,index=[0,1,2,3])
print(df)
```
